```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# methods
def get_rates(column, data):
    y = data[data[column] == 'yes']
    n = data[data[column] == 'no']

    y_y = y[y['y'] == 'yes']
    y_n = y[y['y'] == 'no']

    n_y = n[n['y'] == 'yes']
    n_n = n[n['y'] == 'no']

    y_rate = len(y_y) / len(y_n)
    n_rate = len(n_y) / len(n_n)

    return y_rate, n_rate


def get_rate(column, data, answer):
    x = data[data[column] == answer]

    x_y = x[x['y'] == 'yes']
    x_n = x[x['y'] == 'no']

    x_rate = len(x_y) / len(x_n)

    return x_rate


def get_rate_d(column, data, answer):
    x = data[data[column] != answer]

    x_y = x[x['y'] == 'yes']
    x_n = x[x['y'] == 'no']

    x_rate = len(x_y) / len(x_n)

    return x_rate

def get_pct_yn(column, data):
    y = data[data[column] == 'yes']
    n = data[data[column] == 'no']

    y_y = y[y['y'] == 'yes']
    y_n = y[y['y'] == 'no']

    n_y = n[n['y'] == 'yes']
    n_n = n[n['y'] == 'no']

    y_rate = len(y_y) / len(x)
    n_rate = len(n_y) / len(x)

    return y_rate, n_rate


def get_pct(column, data, answer):
    x = data[data[column] == answer]

    x_y = x[x['y'] == 'yes']
    x_n = x[x['y'] == 'no']

    x_rate = len(x_y) / len(x)

    return x_rate


def get_pct_d(column, data, answer):
    x = data[data[column] != answer]

    x_y = x[x['y'] == 'yes']
    x_n = x[x['y'] == 'no']

    x_rate = len(x_y) / len(x)

    return x_rate
```
